```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("Patient_data7.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
encoder = LabelEncoder()
 
df['Type of Reporting Disease-Condition'] = encoder.fit_transform(df['Type of Reporting Disease-Condition'])    
df['Recent Travel History'] = encoder.fit_transform(df['Recent Travel History'])
df['Any contact with OT case'] = encoder.fit_transform(df['Any contact with OT case'])
df['Presenting Symptom(s), tick if any presented'] = encoder.fit_transform(df['Presenting Symptom(s), tick if any presented'])
df['Underlying Condition(s), tick if any presented'] = encoder.fit_transform(df['Underlying Condition(s), tick if any presented'])
df['Final Laboratory Result'] = encoder.fit_transform(df['Final Laboratory Result'])
```

```python
df.corr()
```

```python
df.head()
```

```python
x = df[['Type of Reporting Disease-Condition','Recent Travel History', 'Any contact with OT case', 'Presenting Symptom(s), tick if any presented', 'Underlying Condition(s), tick if any presented']]
 
y = df[['Final Laboratory Result']]
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.1, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_x, train_y.values.ravel())
```

```python
model.score(test_x, test_y)
```
